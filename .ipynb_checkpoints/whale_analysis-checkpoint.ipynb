```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
whale_returns_df = pd.read_csv(whale_returns_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_returns_df.head()
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df= whale_returns_df.dropna()
whale_returns_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_returns_df.head()
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df= algo_returns_df.dropna()
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_history_df = pd.read_csv(sp500_history_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)

# Sort Index (Date) in ascending order
sp500_history_df= sp500_history_df.sort_index(ascending = True)
sp500_history_df.head()
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types
# Remove $ symbol
sp500_history_df["Close"]= sp500_history_df["Close"].str.replace("$", "")

# convert object type to float for Close
sp500_history_df["Close"]= sp500_history_df["Close"].astype("float")
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
sp500_dailyreturns_df= sp500_history_df.pct_change()
sp500_dailyreturns_df.head()
```

```python
# Drop nulls
sp500_dailyreturns_df= sp500_dailyreturns_df.dropna()
sp500_dailyreturns_df.isnull().sum()
```

```python
# Rename Column
sp500_dailyreturns_df= sp500_dailyreturns_df.rename(columns={"Close": "S&P500"})
sp500_dailyreturns_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_df=pd.concat([whale_returns_df,algo_returns_df, sp500_dailyreturns_df], axis="columns", join="inner")
combined_df.head()
```

# Portfolio Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500?

   **Ans: Based on the Cumulative Returns Plot, it appears as if Berkshire Hathway and Algo 1 have outperformed S&P500.**

```python
# Plot daily returns
combined_df.plot(figsize=(20,10), title="Daily Returns for Whale, Algo and SP500")
```

```python
# Calculate Cumulative Returns
cumulative_returns_df=(1+ combined_df).cumprod()
cumulative_returns_df.head()
```

```python
# Plot Cumulative Returns
cumulative_returns_df.plot(figsize=(20,10), title="Cumulative Returns for Whale, Algo and SP500")
```

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. Which box has the largest spread? Which has the smallest spread?

   **Ans: Berkshire Hathway has the largest spread and Paulson & CO Inc. has the smallest spread.**

2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
   
   **Ans: Berkshire Hathway and Tiger Global are riskier than SP500**
   
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
combined_df.plot.box(figsize=(20,10))
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
daily_std_portfolio= combined_df.std()
daily_std_portfolio
```

```python
# Determine which portfolios are riskier than the S&P 500
daily_std_portfolio > daily_std_portfolio["S&P500"]
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_portfolio_df=daily_std_portfolio*np.sqrt(252)
annualized_portfolio_df
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe_ratio= (combined_df.mean()*252)/annualized_portfolio_df
sharpe_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio.plot(kind="bar")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock. (We have chosen last 2 years worth of historic data)

```python
# Read the first stock - GOOG
goog_csv = Path("Resources/goog_historical.csv")
goog_df = pd.read_csv(goog_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
goog_df.head()
```

```python
# Read the second stock - FB
fb_csv = Path("Resources/fb_historical.csv")
fb_df = pd.read_csv(fb_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
fb_df.head()
```

```python
# Read the third stock - AAPL
aapl_csv = Path("Resources/aapl_historical.csv")
aapl_df = pd.read_csv(aapl_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
aapl_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
combined2_df=pd.concat([goog_df,fb_df, aapl_df], axis="rows", join="inner")
combined2_df.head()
```

```python
# Reset the index
combined2_df=combined2_df.reset_index()
combined2_df.head()
```

```python
# Pivot so that each column of prices represents a unique symbol
combined2_df=combined2_df.pivot(index="Date", columns="Symbol", values="NOCP")
combined2_df.head()
```

```python
# Calculate daily returns
custom_dailyreturns_df= combined2_df.pct_change()
custom_dailyreturns_df.head()
```

```python
# Drop Nulls
custom_dailyreturns_df=custom_dailyreturns_df.dropna()
custom_dailyreturns_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
custom_dailyreturns_df["Custom"]=(custom_dailyreturns_df["AAPL"]+custom_dailyreturns_df["FB"]+custom_dailyreturns_df["GOOG"])/3
custom_dailyreturns_df["Custom"].head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
complete_port_df = pd.concat([combined_df, custom_dailyreturns_df["Custom"]], axis='columns', join='inner')
complete_port_df.head()
```

```python
# check rows
complete_port_df.shape
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
complete_port_df.isnull().sum()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
daily_std_new_portfolio= complete_port_df.std()
daily_std_new_portfolio
```

**Custom portfolio is riskier than S&P500**

```python
# Rolling
complete_port_df.rolling(window=21).std().plot(figsize=(20,10), title= "21 Day Rolling Standard Deviation")
```

```python
# Annualized Sharpe Ratios
sharpe_ratio_new_portfolio= (complete_port_df.mean()*252)/(complete_port_df.std()* np.sqrt(252))
sharpe_ratio_new_portfolio
```

**Custom Portfolio outperforms S&P500**

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio_new_portfolio.plot(kind="bar")
```

```python
# Create a correlation analysis
correlation_new_portfolio = complete_port_df.corr()
correlation_new_portfolio
```

**Custom Portfolio closely mimics Berkshire Hathway and S&P500**

```python
# Rolling Beta - Custom Portfolio

# Calculate rolling variance for S&P 500
rolling_variance_SP500_new = complete_port_df["S&P500"].rolling(window=60).var()

# Calculate rolling Covariance for Custom Portfolio
rolling_covariance_custom= complete_port_df["Custom"].rolling(window=60).cov(complete_port_df["S&P500"])

# Calculate rolling Beta for Custom Portfolio
rolling_beta_custom= rolling_covariance_custom/rolling_variance_SP500_new

# Plot rolling Beta for Custom Portfolio
rolling_beta_custom.plot(figsize=(20,10), title="Custom Portfolio --> Rolling 60 Day Beta")
```
